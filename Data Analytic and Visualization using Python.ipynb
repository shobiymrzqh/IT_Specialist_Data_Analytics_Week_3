```python
yield_apples = [0.89, 0.91, 0.92, 0.93, 0.94, 0.95] #ton per hectare
```

After we define the data then plot line chart using Matloplib

### - Line Chart
A line chart shows data points connected by straight lines, each representing a variable's value at a specific time or category. It's great for displaying trends over time or showcasing variable relationships.

```python
# Create line plot in matloplib
plt.plot(yield_apples)
```

to better understand the graph and its purpose, ad he x-axis values

```python
# Add new list for x-axis 
years = [2017, 2018, 2019, 2020, 2021, 2022]
yield_apples = [0.89, 0.91, 0.92, 0.93, 0.94, 0.95] #ton per hectar
```

parameter 1 consists of an array containing x-axis points, while Parameter 2 comprises an array containing y-axis points.

```python
#plt.plot(x,y)
plt.plot(years, yield_apples)
```

```python
# label the x and y axis
plt.plot(years, yield_apples)
plt.xlabel("Year")
plt.ylabel('Yield (tons per hectare)')
```

To display several datasets on a single graph, employ the plt.plot function separately for each dataset. This is useful when comparing the yields of apples and oranges on a shared graph.

### - Multiple Line Chart
 A type of line chart that displays multiple lines on the same graph to compare trends or relationships between two or more categories.

```python
apples = [0.89, 0.91, 0.93, 0.97, 0.986, 0.943]
tangerines = [0.79, 0.61, 0.93, 0.97, 0.962, 0.931]
```

```python
plt.plot(years, apples)
plt.plot(years, tangerines)
plt.xlabel("Year")
plt.ylabel('Yield (tons per hectare)')
```

This involves generating two line plots through Matplotlib. We can add a legend which tells us what each line in our graph means. To understand what we are plotting, we can add a title to our graph.

```python
plt.plot(years, apples)
plt.plot(years, tangerines)

plt.xlabel("Year")
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Bandung")
plt.legend(['Apples', 'Tangerines'])
```
