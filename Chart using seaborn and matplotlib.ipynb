Using seaborn

```python
apples = [0.89, 0.91, 0.93, 0.97, 0.986, 0.943]
tangerines = [0.79, 0.61, 0.93, 0.97, 0.962, 0.931]
```

```python
plt.plot(years, apples)
plt.plot(years, tangerines)
plt.xlabel("Year")
plt.ylabel('Yield (tons per hectare)')
```

we can add a legend which tells us what each line in our graph means. To understand what we are plotting, we can add a title to our graph.

```python
plt.plot(years, apples)
plt.plot(years, tangerines)

#ngasih nama x & y
plt.xlabel("Year")
plt.ylabel('Yield (tons per hectare)')

#ngasih legend (keterangan line warna A apa,line warna B apa)
plt.title("Crop Yields in Bandung")
plt.legend(['Apples', 'Tangerines'])
```

```python
sns.set_style("whitegrid")
```

```python
plt.plot(years, apples, marker ='o')
plt.plot(years, tangerines, marker ='x')

#ngasih nama x & y
plt.xlabel("Year")
plt.ylabel('Yield (tons per hectare)')

#ngasih legend (keterangan line warna A apa,line warna B apa)
plt.title("Crop Yields in Bandung")
plt.legend(['Apples', 'Tangerines'])
```

```python
#TASK 1
death =[0.2 ,0.25, 0.23, 0.24, 0.30] #persentase kematian
years = [2017, 2018, 2019, 2020, 2021]
grow =[0.5 ,0.4, 0.6, 0.8, 0.4] #persentase pertumbuhan penduduk
```

```python
sns.set_style("whitegrid")
```

```python
plt.plot(years, death)
plt.plot(years, grow)
plt.xlabel("Year")
plt.ylabel('Persentase')

#ngasih legend (keterangan line warna A apa,line warna B apa)
plt.title("Kematian vs Pertumbuhan penduduk ")
plt.legend(['Kematian', 'Pertumbuhan penduduk'])
```

```python
import pandas as pd
```

```python
penduduk = dict(tahun=years, kematian=death, pertumbuhan=grow)
data_penduduk = pd.DataFrame(penduduk)
```

```python
data_penduduk
```

```python
ax = sns.lineplot(x='tahun', y='pertumbuhan', data = data_penduduk)
ax1 = sns.lineplot(x='tahun', y='kematian', data = data_penduduk)

plt.show()
```
